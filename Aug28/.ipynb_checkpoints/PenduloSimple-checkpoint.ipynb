```python
#importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos
from scipy.integrate import odeint


#definimos los parametros
g = 9.81
l = 1

#arreglo de puntos en el tiempo para los cuales resolveremos  
tiempo = np.arange(0, 10.0, 0.1)

#definimos condiciones iniciales
theta_0 = 0.78 #angulo en radianes, aprox 45 grados, definido con ese valor para visualizar mejor la diferencia entre la solucion lineal y la no lineal
x_0 = 0.0 

#definimos una funcion 
def Ec(y,t):
    theta, x = y
    dtheta_dt = [x, -(g/l)*sin(theta)]
    return dtheta_dt

#solucion no lineal
sol_1 = odeint(Ec, [theta_0, x_0], tiempo)

#solucion para angulos pequeños
w = np.sqrt(g/l)
sol_2 = [theta_0 * cos(w*t) for t in tiempo] 

#grafiquemos los resultado, definamos una funcion que grafique
def plot_resultados(tiempo, sol_1, sol_2):
    plt.plot(tiempo, sol_1[:,0])
    plt.plot(tiempo, sol_2)

    plt.title('Pendulo simple')
    plt.xlabel('Tiempo(s)')
    plt.ylabel('Ángulo(rad)')
    plt.legend(['non linear','linear'], loc='upper left')
    plt.show()

plot_resultados(tiempo, sol_1, sol_2)

```
