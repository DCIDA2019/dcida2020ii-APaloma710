```python
#importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos
from scipy.integrate import odeint


#definimos los parametros
g = 9.81
l = 1

#arreglo de puntos en el tiempo para los cuales resolveremos  
tiempo = np.arange(0, 10.0, 0.1)

#definimos condiciones iniciales
theta_0 = 0.78 #angulo en radianes, aprox 45 grados, definido con ese valor para visualizar mejor la diferencia entre la solucion lineal y la no lineal
x_0 = 0.0 

#definimos una funcion 
def Ec(y,t):
    theta, x = y
    dtheta_dt = [x, -(g/l)*sin(theta)]
    return dtheta_dt

#solucion no lineal
sol_1 = odeint(Ec, [theta_0, x_0], tiempo)

#solucion para angulos pequeños
w = np.sqrt(g/l)
sol_2 = [theta_0 * cos(w*t) for t in tiempo] 

#grafiquemos los resultado, definamos una funcion que grafique
def plot_resultados(tiempo, sol_1, sol_2):
    #accesamos solo a la primer columna de sol_1 que corresponde a las soluciones para diferentes tiempos
    plt.plot(tiempo, sol_1[:,0])
    plt.plot(tiempo, sol_2)

    plt.title('Pendulo simple')
    plt.xlabel('Tiempo(s)')
    plt.ylabel('Ángulo(rad)')
    plt.legend(['no lineal','lineal'], loc='upper left')
    plt.show()

plot_resultados(tiempo, sol_1, sol_2)

```

Para poder resolver la ecuación del pendulo simple:

$\frac{\mathrm{d\theta^{2}} }{\mathrm{dt^{2}}} = - \frac{g}{l}\sin(\theta)$

que es una EDO de 2do grado, con el método de Euler debemos primero expresar tal ecuación
como dos ecuaciones de 1er grado acopladas:

$\frac{\mathrm{d}\theta }{\mathrm{d} t} = \omega $


$\frac{\mathrm{d}\omega }{\mathrm{d} t} = - \frac{g}{l} sin(\theta)$

estas deben satisfacer las condiciones inciales:

$\omega_{0} = 0$

$\theta(0) = \theta_{0}$
                
luego, según el método de Euler podemos aproximar $\theta$ y $\omega$ en la forma:

$\theta_{n+1} = \theta_{n} + (h)(\omega_{n}) $

$\omega_{n+1} = \omega_{n} + (h)(f(\theta_{n},t_{n}))$
                          
                    

```python
#importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from math import sin


#parametros
g = 9.81
l = 1

t_0 = 0
t_f = 10
theta_0 = 0.78 #45 grados expresados en radianes
omega_0 = 0
h = 0.01
t = np.arange(t_0,t_f,h)
theta = np.zeros(len(t))
omega = np.zeros(len(t))

theta[0] = theta_0
omega[0] = omega_0
for i in range(len(t)-1):
    omega[i+1] = omega[i] + h*(-(g/l)*sin(theta[i]))
    theta[i+1] = theta[i] + h*omega[i]

plt.plot(t, omega)
plt.plot(t, theta)
plt.title('Pendulo simple con método de Euler')
plt.legend(['omega/(t)','theta(t)'], loc='upper left')
plt.xlabel('Tiempo(s)')
plt.ylabel('Ángulo(rad)')
plt.show()
```
