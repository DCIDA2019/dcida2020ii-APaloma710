Para poder resolver la ecuación del pendulo simple:

$\frac{\mathrm{d\theta^{2}} }{\mathrm{dt^{2}}} = - \frac{g}{l}\sin(\theta)$

que es una EDO de 2do grado, con el método de Euler debemos primero expresar tal ecuación
como dos ecuaciones de 1er grado acopladas:

$\frac{\mathrm{d}\theta }{\mathrm{d} t} = \omega $


$\frac{\mathrm{d}\omega }{\mathrm{d} t} = - \frac{g}{l} sin(\theta)$

estas deben satisfacer las condiciones inciales:

$\omega_{0} = 0$

$\theta(0) = \theta_{0}$
                
luego, según el método de Euler podemos aproximar $\theta$ y $\omega$ en la forma:

$\theta_{n+1} = \theta_{n} + (h)(\omega_{n}) $

$\omega_{n+1} = \omega_{n} + (h)(f(\theta_{n},t_{n}))$

```python
#importamos las librerias necesarias
import matplotlib.pyplot as plt
from math import sin
import time
import sys

#definimos una variable para registrar el tiempo al que inicia la ejecución del código
timestamp1_init = time.time()

#definimos los parametros


def Pendulo(t_fin, n, theta_0, omega_0):
    g = 9.81
    l = 1
    h = t_fin/n
    t = [i*h for i in range(100)]
    omega = [x*0 for x in range(len(t))]
    theta = [j*0 for j in range(len(t))]
    theta[0] = theta_0
    omega[0] = omega_0
    for k in range(len(t)-1):
        theta[k+1] = theta[k] + h*omega[k]
        omega[k+1] = omega[k] + h*(-g/l)*sin(theta[k])
    return theta, omega, t


theta, omega, t = Pendulo(10, 1000, 0.78, 0)       
        
plt.plot(t, omega)
plt.plot(t, theta)
plt.title('Pendulo simple con método de Euler')
plt.legend(['omega(t)','theta(t)'], loc='lower left')
plt.xlabel('Tiempo(s)')
plt.ylabel('Ángulo(rad)')

plt.show()


#guardamos el tiempo final en la variable timestamp1_final.
timestamp1_final = time.time()
#verificamos el tiempo de ejecucion
time1_ex = timestamp1_final - timestamp1_init
print('El tiempo de ejecución fue de: ',time1_ex, 'segundos')
                    
```

```python
#importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
from math import sin
import time
import sys

#definimos una variable para registrar el tiempo al que inicia la ejecución del código
timestamp_init = time.time()

#parametros
g = 9.81
l = 1

def Pendulo(t_0,t_fin, n, theta_0, omega_0):
    g = 9.81
    l = 1
    h = t_fin/n
    t = np.arange(t_0,t_fin,h)
    theta = np.zeros(len(t))
    omega = np.zeros(len(t))
    theta[0] = theta_0
    omega[0] = omega_0
    for i in range(len(t)-1):
        theta[i+1] = theta[i] + h*omega[i]
        omega[i+1] = omega[i] + h*(-g/l)*sin(theta[i])
    return theta, omega, t


theta, omega, t = Pendulo(0,10, 1000, 0.78, 0)


plt.plot(t, omega)
plt.plot(t, theta)
plt.title('Pendulo simple con método de Euler')
plt.legend(['omega/(t)','theta(t)'], loc='upper left')
plt.xlabel('Tiempo(s)')
plt.ylabel('Ángulo(rad)')
plt.show()

timestamp_final = time.time()
#verificamos el tiempo de ejecucion
time_ex = timestamp_final - timestamp_init
print('El tiempo de ejecución fue de: ',time_ex, 'segundos')

diferencia_ex = time1_ex - time_ex
print('La diferencia entre los tiempos de ejecucion es: ', diferencia_ex)
```
