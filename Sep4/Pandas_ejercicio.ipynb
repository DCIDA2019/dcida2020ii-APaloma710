```python
#Importamos las librerias necesarias
#La libreria 'os' provee funciones para interactuar con el sistema operativo
import os
import numpy as np
import pandas as pd
#El modulo "requests" permite enviar solicitudes(delete, get, head) a un url especifico
import requests

file = 'salarios_minimos.csv'
if not os.path.isfile(file):
    url = "https://idconline.mx/laboral/salarios-minimos/salarios-minimos-1986-2007"
    html = requests.get(url).content
    df_list = pd.read_html(html, header = 1)
    df = df_list[-1]
    df.to_csv('salarios_minimos.csv')
else:
    df = pd.read_csv(file)
    
df
```
